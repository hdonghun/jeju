```python
def parse_coordinates(coord_str):
    coordinates = {}
    lines = coord_str.strip().split('\n')
    
    for line in lines:
        key, value = line.split(':')
        lat, lon = map(float, value.split(','))
        coordinates[key] = (lat, lon)
    
    return coordinates

def parse_coordinates_from_file(file_path):
    with open(file_path, 'r') as file:
        coord_str = file.read()
    return parse_coordinates(coord_str)

def calculate_overlap_and_gap(coordinates_list, file_names):
    result_messages = []

    for i, coordinates in enumerate(coordinates_list):
        all_latitudes = [coord[0] for coord in coordinates.values()]
        all_longitudes = [coord[1] for coord in coordinates.values()]

        min_lat, max_lat = min(all_latitudes), max(all_latitudes)
        min_lon, max_lon = min(all_longitudes), max(all_longitudes)

        lat_gap = max_lat - min_lat
        lon_gap = max_lon - min_lon

        if lat_gap < lon_gap:
            result_messages.append(f"{file_names[i]}: 겹침이 있습니다.")
        elif lon_gap < lat_gap:
            result_messages.append(f"{file_names[i]}: 겹침이 있습니다.")
        else:
            result_messages.append(f"{file_names[i]}: 간격이 있습니다. 위도(Latitude) 간격: {lat_gap}, 경도(Longitude) 간격: {lon_gap}")

    return result_messages

# 파일들로부터 좌표를 읽어들이기
coordinates_list = [parse_coordinates_from_file(file_path) for file_path in cdn_path_list]

# 겹침 및 간격 계산
result_messages = calculate_overlap_and_gap(coordinates_list, cdn_path_list)

# 결과 출력
for result in result_messages:
    print(result)
```
