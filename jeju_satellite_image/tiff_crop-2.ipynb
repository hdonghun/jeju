```python
# 색 구별 확실히 하기 
import cv2
import numpy as np

# 컬러 이미지를 불러옵니다.
img = cv2.imread('/home/alpaco/jeju/output.tif')

# 이미지를 YUV 컬러 공간으로 변환합니다.
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

# Y 채널에 대해 히스토그램 평활화를 적용합니다.
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])

# 이미지를 다시 RGB 컬러 공간으로 변환합니다.
img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

# 결과를 저장합니다.
cv2.imwrite('/home/alpaco/jeju/output-2.tif', img_output)
```

```python
# 색 구별 확실히 한 파일 시각화 
color_image = cv2.imread('/home/alpaco/jeju/output-2.tif', cv2.IMREAD_COLOR)


print(color_image.shape)
# 이미지 표시
plt.imshow(color_image)
plt.show()
print(color_image[0])
```
